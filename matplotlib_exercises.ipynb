```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import math
from random import randint
import random
```

```python
# Use matplotlib to plot the following equation:

#y=x2−x+2
# You'll need to write the code that generates the x and y points.

# Add an anotation for the point 0, 0, the origin.

x = np.arange(50)
y = (x*x)-x+2

plt.xlabel('x')
plt.ylabel('y')
plt.title('Equation')
plt.annotate('origin', xy = (0, 0),xytext=(0, 0), arrowprops = dict(facecolor = 'gray')  )


plt.plot(x,y)
plt.show()
```

```python
# Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# y=√x
# y=x3
# y=2x
# y=1/(x+1)

# You can use functions from the math module to help implement some of the equations above.

# y=√x
x = np.array([0, 1, 2, 3, 4])
y1 = np.zeros(len(x))

for i in range(len(x)):
    y1[i] = np.sqrt(x[i])

plt.xlabel('x')
plt.ylabel('sqaure root')
plt.title('square root of x')

plt.plot(x,y1)
plt.show()
```

```python
# y=x3
x = np.array([0, 1, 2, 3, 4])
y2 = x**3

plt.xlabel('x')
plt.ylabel('cubed')
plt.title('x cubed')

plt.plot(x,y2)
plt.show()
```

```python
# y=2x
x = np.array([0, 1, 2, 3, 4])
y3 = x *2

plt.xlabel('x')
plt.ylabel('doubled')
plt.title('x doubled')

plt.plot(x,y3)
plt.show()
```

```python
# y=1/(x+1)
x = np.array([0, 1, 2, 3, 4])
y4 = 1/(x+1)

plt.xlabel('x')
plt.ylabel('y')
plt.title('1/(x+1)')

plt.plot(x,y4)
plt.show()
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.

plt.figure(figsize = (8,2))

plt.subplot(2, 2, 1)
plt.text(0.5, 0.5, str((221)))
plt.scatter(x, y)

plt.subplot(222)
plt.text(0.5, 0.5, str((222)))
plt.plot(x, y)

plt.subplot(223)
plt.text(0.5, 0.5, str((223)))
plt.plot(x, y)

plt.subplot(224)
plt.text(0.5, 0.5, str((224)))
plt.plot(x, y)
```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations 
# has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
y1 = np.zeros(len(x))
for i in range(len(x)):
    y1[i] = math.sqrt(x[i])
y2 = x**3
y3 = x *2
y4 = 1/(x+1)

plt.figure(figsize = (8,8))
plt.suptitle('Some Math Functions')

plt.subplot(221)
plt.plot(x, y1,color = 'r', ls = '--')
plt.title('square root')

plt.subplot(222)
plt.plot(x, y2,color = 'r', ls = '--', marker = '^')
plt.title('cubed')

plt.subplot(223)
plt.plot(x, y3,color = 'g', ls = '--', marker = 'o')
plt.title('doubled')

plt.subplot(224)
plt.plot(x, y4,color = 'm', ls = '--', marker = 'v')
plt.axhline(0, color = 'm')
plt.fill_between(x, y4, color = 'm', alpha= 0.9)
plt.title('1/(x+1)')

plt.show()
```
